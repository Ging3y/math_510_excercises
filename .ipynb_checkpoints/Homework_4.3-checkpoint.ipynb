```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
def plot_vecs(A):
    
    U, S, V = la.svd(A)
    
    sig_1 = S[0]
    sig_2 = S[1]
    
    u_1 = U[:,0]*sig_1
    u_2 = U[:,1]*sig_2
    
    v_1 = V[:,0]
    v_2 = V[:,1]
        
    plt.figure(figsize=(20,10))
    plt.suptitle(str(A), fontsize=25)
    plt.subplot(121)
    plt.title("Right Singular Vectors", fontsize=15)
    t = np.linspace(0, 2*np.pi, 250)
    plt.plot(np.cos(t), np.sin(t), linewidth=3)
    origin = ([0], [0]) # origin point
    plt.quiver(*origin, *v_1, angles='xy', scale_units='xy', scale=1)
    plt.quiver(*origin, *v_2, angles='xy', scale_units='xy', scale=1)
    plt.subplot(122)
    plt.title("Left Singular Vectors", fontsize=15)
    DOMAIN = np.linspace(0, 2*np.pi, 250)
    transform = np.array([A@np.array([np.cos(t), np.sin(t)]) for t in DOMAIN])
    plt.plot(transform[:,0], transform[:,1], linewidth=3)
    plt.quiver(*origin, *u_1, angles='xy', scale_units='xy', scale=1)
    plt.quiver(*origin, *u_2, angles='xy', scale_units='xy', scale=1)
    plt.show()
```

```python
A = [np.array([[1,2],[0,2]]), np.array([[3,0],[0,-2]]), np.array([[0,2],[0,0],[0,0]]), np.array([[1,1],[0,0]]), np.array([[1,1],[1,1]])]

for m in A:
    plot_vecs(m)
```
